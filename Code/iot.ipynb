```python
import qrcode
import os

#df = pd.read_csv('cleaned_ids2018_sampled.csv')


# Exclude the last column
df_except_last = sampled_df.iloc[:, :-1]

print("Original sampled_df shape:", sampled_df.shape)
print("New df_except_last shape:", df_except_last.shape)

# Directory to save QR codes
output_directory = 'qr_codes2'
os.makedirs(output_directory, exist_ok=True)

# Define QR code parameters
#qr_version = 1
#qr_error_correction = qrcode.constants.ERROR_CORRECT_L
#qr_box_size = 10
#qr_border = 4
qr_width = 1410
qr_height = 1410

for index, row in enumerate(df_except_last.itertuples(), start=1):
    # Create a string representation of the row
    row_data = str(row[1:])  # Exclude the first element which is the index

    # Create a QR code for the row data
    qr = qrcode.QRCode()
    qr.add_data(row_data)
    qr.make(fit=True)

    # Create an image from the QR code with specific dimensions and one-bit depth
    img = qr.make_image(fill_color="black", back_color="white")
    img = img.resize((qr_width, qr_height))
    img = img.convert('1')  # Convert to one-bit depth

    # Save the QR code image with a unique filename
    filename = os.path.join(output_directory, f'qrcode_{index}.png')
    img.save(filename, format='PNG')  # Set dpi if needed

    print(f"QR code saved: {filename}")
```

```python
import os

# Extract the last column
last_column = sampled_df['Label']

# Define QR code parameters
qr_width = 1410
qr_height = 1410

# Directory to save QR codes
output_directory = 'output_qr_codes'
os.makedirs(output_directory, exist_ok=True)

# Iterate through each row in the dataset
for index, label_value in enumerate(last_column, start=1):
    # Create a QR code for the label value
    qr = qrcode.QRCode()
    qr.add_data(str(label_value))
    qr.make(fit=True)

    # Create an image from the QR code with specific dimensions and one-bit depth
    img = qr.make_image(fill_color="black", back_color="white")
    img = img.resize((qr_width, qr_height))
    img = img.convert('1')  # Convert to one-bit depth

    # Save the QR code image with a unique filename
    filename = os.path.join(output_directory, f'qrcode_label_{label_value}_{index}.png')
    img.save(filename, format='PNG')  # Set dpi if needed

    print(f"QR code saved: {filename}")
```
